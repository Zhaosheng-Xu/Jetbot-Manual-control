# 🕹️ JetBot 手动控制（使用 traitlets.dlink）

本 notebook 使用 traitlets 的 dlink 方法将游戏手柄轴绑定到 JetBot 的左右电机，实现基础移动控制。

```python
# 导入库
from jetbot import Robot
import ipywidgets.widgets as widgets
from IPython.display import display
import traitlets
```

```python
# 初始化 JetBot 和手柄
robot = Robot()
controller = widgets.Controller(index=0)
display(controller)
```

```python
# 将左右摇杆绑定到左右电机（通常为 axes[1] 和 axes[3]）
left_link = traitlets.dlink((controller.axes[1], 'value'),
                            (robot.left_motor, 'value'),
                            transform=lambda x: -x)

right_link = traitlets.dlink((controller.axes[3], 'value'),
                             (robot.right_motor, 'value'),
                             transform=lambda x: -x)
```

```python
# 添加停止按钮（紧急停止）
stop_button = widgets.Button(description='紧急停止', button_style='danger')
display(stop_button)

def stop_robot(b):
    robot.stop()

stop_button.on_click(stop_robot)
```
